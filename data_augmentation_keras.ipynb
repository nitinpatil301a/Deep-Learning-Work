```
datagen=ImageDataGenerator(rotation_range=40,
                           zoom_range=0.2,
                           vertical_flip=True,
                           horizontal_flip=True,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           shear_range=0.2)

img=load_img('/content/Tensorflow-Keras-Work/cat.jpg',target_size=(150,150,3))
```

```
x=img_to_array(img)
x=np.reshape(x,(1,150,150,3))
```

```
x.shape
```

```
i=0
for batch in  datagen.flow(x,save_to_dir='/content/Tensorflow-Keras-Work/',save_prefix='cat',):
  print(batch)
  i+=1
  if(i==10):
    break
```
